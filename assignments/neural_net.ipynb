```python
class Layer:
    def __init__(self,):
        pass
    def forward(self, input_x):
        pass

class Linear:
    def __init__(self, input_len, output_len):
        self.W = np.ones((output_len, input_len))
        self.b = np.zeros(output_len)
    
    def forward(self, input_x):
        return self.W @ input_x + self.b
    
    def get_nparams(self):
        return self.W.shape[0]*self.W.shape[1] + len(self.b)
    
class ReLU():
    def forward(self, input_x):
        return np.maximum(input_x, 0)

class Softmax():
    def forward(self, input_x):
        return input_x
    
class Conv():
    def __init__(self, kernel_size, cin, cout, padding, stride):
        pass
    def forward(self, input_x):
        pass

class Flatten():
    pass
    
class NeuralNetwork:
    def __init__(self, layers):
        self.layers = layers
    
    def forward(self, input_x):
        x = input_x
        for layer in layers:
            x = layer.forward(x)
        return x
```

```python
input_x = np.ones(784)
```

```python
layers = [Linear(784, 100),
          ReLU(),
          Linear(100, 10),
          Softmax()]

net = NeuralNetwork(layers)

net.forward(input_x)
```
